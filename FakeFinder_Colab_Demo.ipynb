```python
# 📁 Upload image
from google.colab import files
from PIL import Image
import cv2
import numpy as np
import io

uploaded = files.upload()
image = None

for filename in uploaded.keys():
    image = Image.open(io.BytesIO(uploaded[filename]))
    image.save("uploaded_image.jpg")
    print(f"Saved {filename} as uploaded_image.jpg")
```

```python
# 🧠 Simulated FakeFinder Detection
import random

def fakefinder_detect(image_path):
    # Simulate prediction with a random score
    confidence = round(random.uniform(0.6, 0.99), 2)
    label = "Deepfake" if confidence > 0.8 else "Real"
    return label, confidence

label, confidence = fakefinder_detect("uploaded_image.jpg")
print(f"Prediction: {label}\nConfidence: {confidence * 100}%")
```
